```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many

```python
# 1. Report the account representative for each customer
run("""
SELECT 
    customerName, CONCAT(lastname, ', ', firstname) AS employeeName
FROM
    customers
        INNER JOIN
    employees 
ON salesRepEmployeeNumber = employeeNumber
""")
```

```python
# 2. Report total payments for Atelier graphique

run("""
SELECT SUM(amount) AS Total_Pay
FROM payments INNER JOIN customers
USING (customerNumber)
WHERE customerName='Atelier graphique'
""")
```

```python
# 3. Report the total payments by date

run("""
SELECT paymentDate, SUM(amount) AS Total_Pay
FROM payments
GROUP BY paymentDate
""")
```

```python
# 4. Report the products that have not been sold

run("""
SELECT productName
FROM products a LEFT JOIN orderdetails b
ON a.productCode = b.productCode
WHERE b.productCode is NULL
""")
```

```python
# 5. List the amount paid by each customer

run("""
SELECT customerName, SUM(amount) AS total_paid
FROM customers INNER JOIN payments
USING(customerNumber)
GROUP BY customerNumber
""")
```

```python
# 6. How many orders have been placed by Herkku Gifts?

run("""
SELECT customerName, COUNT(*) AS JmlOrder
FROM orders c INNER JOIN customers d
ON c.customerNumber = d.customerNumber
WHERE customerName = 'Herkku Gifts'
""")
```

```python
# 7. Who are the employees in Boston?
run("""
SELECT CONCAT(firstName,' ',lastName) AS employeeName, city
FROM employees INNER JOIN offices
USING(officeCode)
WHERE city = 'Boston'
""")
```

```python
# 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.

run("""
SELECT customerName, amount
FROM customers INNER JOIN payments
USING (customerNumber)
WHERE amount > 100000
ORDER BY amount DESC
""")
```

```python
# 9. List the value of 'On Hold' orders.

run("""
SELECT SUM(quantityOrdered*priceEach) AS Nilai
FROM orderdetails INNER JOIN orders
USING(orderNumber)
WHERE status = 'On Hold'
""")
```

```python
# 10. Report the number of orders 'On Hold' for each customer.

run("""
SELECT customerName, COUNT(*) AS NumOnHold
FROM customers INNER JOIN orders
USING(customerNumber)
WHERE status = 'On Hold'
GROUP BY customerNumber
""")
```

# Many to many

```python
# 1. List products sold by order date.

run("""
SELECT productName, orderDate
FROM products
INNER JOIN orderdetails USING(productCode)
INNER JOIN orders USING(orderNumber)
ORDER BY orderDate
""")
```

```python
# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

run("""
SELECT orderNumber,orderDate
FROM products
INNER JOIN orderdetails USING(productCode)
INNER JOIN orders USING(orderNumber)
WHERE productName = '1940 Ford Pickup Truck'
ORDER BY orderDate DESC
""")
```

```python
# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

run("""
SELECT customerName, orderNumber, SUM(quantityOrdered*priceEach) AS jumlah
FROM customers
INNER JOIN orders USING(customerNumber)
INNER JOIN orderdetails USING(orderNumber)
GROUP BY orderNumber
HAVING jumlah > 25000
""")
```

```python
# 4. Are there any products that appear on all orders?

run("""
SELECT productCode, COUNT(*) AS jml
FROM orderdetails
GROUP BY productCode
HAVING jml = (SELECT COUNT(*) FROM orders)
""")
```

```python
# 5. List the names of products sold at less than 80% of the MSRP.

run("""
SELECT orderNumber, productName, priceEach, MSRP*80/100 AS eightyMSRP
FROM products INNER JOIN orderdetails
USING (productCode)
HAVING priceEach < eightyMSRP
""")
```

```python
# 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

run("""
SELECT productName, priceEach-buyPrice AS adv, buyPrice
FROM products INNER JOIN orderdetails USING(productCode)
WHERE  priceEach-buyPrice > buyPrice
""")
```

```python
# 7. List the products ordered on a Monday.

run("""
SELECT DAYNAME(orderDate) AS day, productName
FROM products INNER JOIN orderdetails USING(productCode)
INNER JOIN orders USING(orderNumber)
WHERE DAYOFWEEK(orderDate)=2
""")
```

```python
# 8. What is the quantity on hand for products listed on 'On Hold' orders?

run("""
SELECT SUM(quantityOrdered) AS jmlProd
FROM orders INNER JOIN orderdetails USING(orderNumber)
WHERE status = 'On Hold'
""")
```
